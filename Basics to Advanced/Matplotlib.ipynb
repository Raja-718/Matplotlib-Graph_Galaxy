#**Bar Chart**

```python
# simple bar chart
children = [25,20,40,15,30]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors, children)
```

```python

# horizontal bar chart
children = [25,20,40,15,30]
colors = ['red','blue','green','yellow','pink']
plt.barh(colors, children)
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
```

```python
df.head(5)
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'],width = 0.6)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'],width = 0.2)
plt.bar(np.arange(df.shape[0]), df['2016'],width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2016'],width = 0.2)


```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'],width = 0.2,color = '#FA7B05')
plt.bar(np.arange(df.shape[0]), df['2016'],width = 0.2,color = '#F4F2F1')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2016'],width = 0.2,color = '#02971B')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python

# horizontal bar chart
children = [25,20,40,15,30]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors, children)
plt.xticks(rotation ='vertical')
```

```python
# stack bar chart
plt.bar(df['batsman'],df['2017'],label = '2017')
plt.bar(df['batsman'],df['2016'],bottom= df['2017'],label = '2016')
plt.bar(df['batsman'],df['2015'],bottom = (df['2016'] + df['2017']),label = '2015')
plt.legend()
plt.show()
```
