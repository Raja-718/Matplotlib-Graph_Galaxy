```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c =iris['Species'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
scatter = plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],
                      c=iris['Species'].astype('category').cat.codes, cmap='jet')

plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')

plt.colorbar(scatter, label='Species')
plt.show()
```

```python
# increase or decrese size of graph
plt.figure(figsize = (10,7))
scatter = plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],
                      c=iris['Species'].astype('category').cat.codes, cmap='jet')

plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')

plt.colorbar(scatter, label='Species')
plt.show()
```

#**Annotations**

```python
batters = pd.read_csv('/content/batter.csv')
```

```python
batters.shape
```

```python
batters.head(5)
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize = (18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s = sample_df['runs'])
for i  in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

#**Horizontal and Vertical Lines**

```python
plt.figure(figsize = (18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s = sample_df['runs'])
plt.axhline(130,color ='red')
plt.axvline(30,color ='red')
for i  in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

**3D Surface plots**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

**Contour Plots**

```python
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(projection='3d')
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(projection='3d')
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
